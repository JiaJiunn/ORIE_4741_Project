### Preprocessing
I will use the county data and augment it to include things such as election results, latitude longitude, state

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
county_2012 = pd.read_csv("./US-county-demos/county_stats_2012.csv", encoding = 'unicode_escape')
county_2012.head()
```

```python
###data types of all columns
county_2012.dtypes
```

```python
county_2012['County']
```

```python
#add state column
state = []
for county in county_2012['County']:
    try:
        state.append(str(county).split(',')[1].strip())
    except IndexError:
        state.append(None)

state[0:20]

```

```python
county_2012['State']=state
```

```python
county_2012.head()
```

```python
#Just to see what a NaN county looks like
d=county_2012[county_2012['State'].isnull()]
d
```

```python
##Seems to be an actual county 
#quick google search (https://mn.postcodebase.com/county/27165) reveals this is Watonwan County, MN
county_2012.at[1396,'County'] = 'Watonwan County, MN'
county_2012.at[1396, 'State'] = 'MN'
county_2012.iloc[1396,0:11]
```

```python
#clean up county column
county_2012['County'] = [county.split(',')[0].strip() for county in county_2012['County']]
```

```python
county_2012_rev = county_2012.iloc[:,0:11]
county_2012_rev
```

```python
#dispose of all non-voting counties (i.e sry Puerto Rico)
states_list=["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]

county_2012_rev2 = county_2012_rev[county_2012_rev['State'].isin(states_list)]
county_2012_rev2
```

```python
###integrate geographic data
geo_data = pd.read_csv('./Geocodes_USA_with_Counties.csv')
```

```python
geo_data.head()
```

```python
geo_data['County'] = [str(county) + ' County' for county in geo_data['county']]
geo_data
```

```python
county_2012_rev2['County'] = county_2012_rev2['County'].astype(str)
county_2012_rev2.dtypes
```

```python
geo_data['County'] = geo_data['County'].astype(str)
```

```python
county_2012_join = pd.merge(county_2012_rev2,geo_data, on="County", how="left")
county_2012_join
```

```python
county_2012_join.dtypes
```

```python
cols_to_keep = ['FIPS', 'County', 'DEM', 'GOP', 'MedianIncome', 'MigraRate', 'BirthRate', 'DeathRate', 'BachelorRate', 
                'UnemploymentRate', 'State', 'zip','primary_city', 'latitude', 'lognitude', 'estimated_population']
county_2012_final=county_2012_join.filter(items=cols_to_keep, axis='columns')
county_2012_final
```

```python
#Remove duplicate columns (1 county with multiple zip codes)
county_2012_final2 = county_2012_final.drop_duplicates(subset=['FIPS'])
county_2012_final2
```

```python
county_2012_final2.to_csv(path_or_buf="./county_2012_preprocessed.csv",index=False)
```
