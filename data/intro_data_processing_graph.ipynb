```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
```

```python
graph_df = pd.read_csv("./US-county-demos/county_graph.csv", encoding = 'unicode_escape')
graph_df
```

```python
graph_df.dtypes
```

```python
#subset based on our previously processed data
county_data = pd.read_csv("./county_2012_preprocessed.csv")
county_data
```

```python
for i in county_data['FIPS']:
    print(type(i))
```

```python
idx_final = []
county_fips = list(county_data.loc[:,'FIPS'])

for i in range(len(graph_df['SRC'])):
    src = int(graph_df.loc[i,'SRC'])
    dst = int(graph_df.loc[i,'DST'])
    if (src in county_fips) and (dst in county_fips):
        idx_final.append(i)
```

```python
graph_final = graph_df.iloc[idx_final,:]
graph_final
```

```python
graph_final.to_csv(path_or_buf="./county_graph_preprocessed.csv",index=False)
```
