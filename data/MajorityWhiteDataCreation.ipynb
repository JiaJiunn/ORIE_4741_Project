```python
###Generates whether a county is majority white
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#load data
df = pd.read_csv("./cc-est2012-alldata.csv", encoding="latin-1")
df.head()
```

```python
#Remove extra columns
df1=df.filter(items=['STNAME','CTYNAME', 'YEAR','AGEGRP','TOT_POP','WA_MALE','WA_FEMALE'])
df1.head()
```

```python
df2 = df1.query('YEAR == 5 & AGEGRP == 0')
df2.head()
```

```python
df2['Majority_White'] = [(df2.iloc[i,5]+df2.iloc[i,6])/df2.iloc[i,4]> 0.5 for i in range(len(df2)) ]
df2['Majority_White'][:50]
```

```python
sum(df2['Majority_White'])/len(df2['Majority_White'])
```

```python
#^^Yikes
df2['Percent_White'] = [(df2.iloc[i,5]+df2.iloc[i,6])/df2.iloc[i,4] for i in range(len(df2)) ]
df2.head()
```

```python
df_final = df2.filter(items=['STNAME','CTYNAME','Majority_White','Percent_White'])
df_final.head()
```

```python
df_final.to_csv("majority_white2012.csv", index=False)
```
