```python
def count_nulls(col):
    count=0
    for i in col:
        if pd.isna(i):
            count += 1
    return count
```

```python
count_nulls(df_2012['latitude'])
```

```python
count_nulls(df_2012['longitude'])
```

```python
def idx_nulls(col):
    idx_lst=[]
    for i in range(len(col)):
        if pd.isna(col[i]):
            idx_lst.append(i)
    return idx_lst
```

```python
df_2012.loc[idx_nulls(df_2012['latitude']),:]
```
